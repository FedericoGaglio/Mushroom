## Feature selection

Per andare a semplicare il modello ed evitare un eccessivo overfitting vado ad utilizzare, per la scelta di un sottoinsieme dei vari attributi, **SelectFromModel**.

Questo strumento, offerto da sklearn, è di fatto un meta-trasformatore che a seconda dello stimatore che gli viene passato, va a fare una selezione delle features ad-hoc.

In questo caso ho scelto di utilizzare **Lasso**, che è un ottimo strumento per poter andare a diminuire l'overfitting.

Lasso può essere definita come segue:

> $$L(\boldsymbol{w}) = \sum_{i=1}^{N}(y_i - f(x_i))^2 +\lambda R(\boldsymbol{w})$$

>> con fattore di regolarizzazione:
>> $$R(\boldsymbol{w}) = \sum_l \left | w_l \right |$$

Come si può notare nella LASSO, il modello penalizza tutti i parametri indifferentemente, schiacciandone alcuni a 0. Questo ha l’effetto collaterale di eliminare alcune features, semplificando il modello.
 

```python
from sklearn.feature_selection import SelectFromModel
from sklearn.linear_model import Lasso


sfm=SelectFromModel(Lasso(max_iter=100000, alpha=0.01))
sfm.fit(X, y)

sfm.get_support()

selected_Lasso= X.columns[(sfm.get_support())]


#len(selected_Lasso)
selected_Lasso
```

```python
X = X[selected_Lasso]
X
```

## Addestramento del modello

Come classificatore ho scelto come si può notare il RandomForestClassifier. La scelta è ricaduta su di esso in quanto presenta molte caratteristiche e vantaggi interessanti. In generale l'idea di base è la seguente:
i random forest creano alberi decisionali su campioni di dati selezionati casualmente, ottengono previsioni da ciascun albero e selezionano la soluzione migliore mediante votazione.

Quindi riassumendo in punti quale è il suo funzionamento abbiamo che:

> * seleziona randomicamente i samples da un dataset fornito
> * costruisce un decision tree per ogni sample e va a prendere il risultato della predizione da ogni decision tree
> * assegna un voto per ogni risultato previsto
> * seleziona il risultato della previsione con il maggior numero di voti come previsione finale. 

Per quanto riguarda i vantaggi che si hanno andando ad utilizzare RandomForest sicuramente il più importante è che non soffre di problemi di overfitting.

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,random_state=0)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier()

#fit e prediction
rf.fit(X_train, y_train)
pred = rf.predict(X_valid)
print('Accuracy:',accuracy_score(y_valid, pred))

```
