```python
from sklearn.preprocessing import LabelEncoder


label_encoder = LabelEncoder()

for i in dataset:
    dataset[i] = label_encoder.fit_transform(dataset[i])
    
dataset
```

```python
dataset.class_edible.unique()
```

```python
y = dataset.class_edible

X = dataset.drop(columns=['class_edible'])
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2,random_state=0)
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
  
rf = RandomForestClassifier(criterion='gini',
                            n_estimators=1750,
                            max_depth=7,
                            min_samples_split=6,
                            min_samples_leaf=6,
                            max_features='auto',
                            oob_score=True,
                            random_state=42,
                            n_jobs=-1,
                            verbose=0)



sv = SVC(probability=True, random_state=42)
logreg = LogisticRegression(n_jobs=-1, solver='newton-cg', random_state=42)
gb = GradientBoostingClassifier(random_state=42)
gnb = GaussianNB()
mlp = MLPClassifier(random_state=42)
 
estimators = [rf, gb, sv, logreg, gnb, mlp]
```

```python
from sklearn.metrics import accuracy_score


for estimator in estimators:
    estimator.fit(X_train, y_train)
    preds = estimator.predict(X_valid)
    print(estimator)
    print('Accuracy:',accuracy_score(y_valid, preds))
    print('\n')
    print('-----------------------------------------------')
```
